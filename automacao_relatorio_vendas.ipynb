# Automação de Relatório de Vendas com Python

### Contexto do Problema

Todos os dias, o sistema da empresa atualiza as vendas do dia anterior. 
A tarefa do analista é enviar um e-mail para a diretoria com os principais indicadores: 
**faturamento total** e **quantidade de produtos vendidos**.

Este projeto automatiza todo esse fluxo — da coleta dos dados até o envio do e-mail — eliminando o trabalho manual diário.

---

### Fluxo da Solução

1. **Acesso ao sistema** — abertura do navegador e navegação até a fonte de dados
2. **Extração** — download automático da base de dados em Excel
3. **Processamento** — leitura e cálculo dos indicadores com Pandas
4. **Entrega** — geração e envio automático do relatório por e-mail

---

**Biblioteca de automação:** [pyautogui](https://pyautogui.readthedocs.io/en/latest/quickstart.html)

```python
# Dependências do projeto
# Instalação: pip install pandas pyautogui openpyxl pyperclip

# pyautogui  -> automação de mouse e teclado
# pandas     -> leitura e processamento de dados
# openpyxl   -> suporte a arquivos Excel (.xlsx)
# pyperclip  -> manipulação da área de transferência

# Principais métodos do pyautogui utilizados:
# pyautogui.click()   -> clique do mouse em coordenadas (x, y)
# pyautogui.write()   -> digitação de texto
# pyautogui.press()   -> acionamento de tecla
# pyautogui.hotkey()  -> combinação de teclas
# pyautogui.position()-> retorna a posição atual do cursor
```

```python
import pyautogui
import time

pyautogui.pause = 2  # intervalo de segurança entre ações

# ── 1. ACESSO AO SISTEMA ────────────────────────────────────────────────────
# Abre o navegador via menu iniciar e navega até a fonte de dados

pyautogui.press("win")
pyautogui.write("chrome")
pyautogui.press("enter")
time.sleep(5)

link = "LINK_DO_SISTEMA"  # substituir pelo endereço real da fonte de dados
pyautogui.write(link)
pyautogui.press("enter")
time.sleep(4)
```

```python
# ── 2. EXTRAÇÃO DA BASE DE DADOS ────────────────────────────────────────────
# Navega até o arquivo e realiza o download via menu de contexto

pyautogui.click(x=603, y=467, clicks=2, interval=0.25)  # abre a pasta
time.sleep(2)

pyautogui.click(x=444, y=456, button="right")  # abre menu de contexto no arquivo
time.sleep(2)
pyautogui.click(x=537, y=556)  # seleciona a opção de download
time.sleep(5)

# Nota: as coordenadas (x, y) variam conforme a resolução da tela.
# Use pyautogui.position() para mapear os pontos no seu ambiente.
```

### Processamento dos Indicadores

Com o arquivo baixado, realizamos a leitura com Pandas e calculamos:
- **Faturamento total** — soma da coluna `Valor Final`
- **Quantidade de produtos vendidos** — soma da coluna `Quantidade`

```python
import os
import pandas as pd

# ── 3. PROCESSAMENTO DOS DADOS ───────────────────────────────────────────────
# Identifica automaticamente o arquivo Excel mais recente na pasta de downloads

pasta_downloads = os.path.join(os.path.expanduser("~"), "Downloads")
arquivos = [os.path.join(pasta_downloads, f) for f in os.listdir(pasta_downloads) if f.endswith(".xlsx")]
arquivo_recente = max(arquivos, key=os.path.getctime)

tabela = pd.read_excel(arquivo_recente)
display(tabela)

# Cálculo dos indicadores
faturamento    = tabela["Valor Final"].sum()
qtde_produtos  = tabela["Quantidade"].sum()

print(f"Faturamento total: R$ {faturamento:,.2f}")
print(f"Qtde de produtos vendidos: {qtde_produtos:,}")
```

### Geração e Envio do Relatório

Com os indicadores calculados, o relatório é formatado e enviado por e-mail via Gmail.

```python
import pyperclip

# ── 4. ENVIO DO RELATÓRIO POR E-MAIL ────────────────────────────────────────
# Abre o Gmail, preenche destinatário, assunto e corpo, e envia

pyautogui.hotkey("ctrl", "t")
pyautogui.write("mail.google.com")
pyautogui.press("enter")
time.sleep(5)

pyautogui.click(x=55, y=276)  # botão "Escrever"
time.sleep(3)

# Destinatário
pyautogui.write("diretoria@empresa.com")  # substituir pelo e-mail real
time.sleep(1.5)
pyautogui.press("tab")
pyautogui.press("tab")

# Assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")

# Corpo do e-mail com os indicadores calculados
texto = f"""
Prezados,

Segue o relatório de vendas de hoje.

Faturamento: R$ {faturamento:,.2f}
Quantidade de produtos vendidos: {qtde_produtos:,}

Qualquer dúvida estou à disposição.

Att,
Analista de Dados
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# Envio
pyautogui.hotkey("ctrl", "enter")
```
